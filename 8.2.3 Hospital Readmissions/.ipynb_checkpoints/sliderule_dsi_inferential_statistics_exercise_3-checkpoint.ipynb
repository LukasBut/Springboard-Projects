# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

### My Code Solution (statistical analysis)

```python
#Following on from previous data cleaning steps I keep only the relevant columns and drop rows with NaN values.
clean_hospital_read_df=clean_hospital_read_df.loc[:,["Number of Discharges","Excess Readmission Ratio",
                                                     "Predicted Readmission Rate", "Expected Readmission Rate",
                                                     "Number of Readmissions","Start Date", "End Date"]].dropna()
#Correcting the "Number of Readmissions" column variable type, float->int.
clean_hospital_read_df["Number of Readmissions"]=clean_hospital_read_df["Number of Readmissions"].astype(int)
#We must reset the index after having deleted NaN valued rows.
clean_hospital_read_df=clean_hospital_read_df.reset_index()

#I needed the pearsonr function from scipy.stats to calculate the correlation coefficient.
from scipy import stats

#Computing sample statistics
x_mean=np.mean(clean_hospital_read_df["Number of Discharges"])
y_mean=np.mean(clean_hospital_read_df["Excess Readmission Ratio"])
x_std=np.std(clean_hospital_read_df["Number of Discharges"])
y_std=np.std(clean_hospital_read_df["Excess Readmission Ratio"])

print("Number of Discharges mean and std: (%f, %f) , Readmission Ratio mean and std: (%f, %f)" % (x_mean,x_std,y_mean,y_std))
```

After exploring the *hospital_read_df* dataframe I continued with a few more data cleaning steps to ensure I wouldn't have to worry about data quality issues later on in my analysis. Rows with NaN values made up a very small proportion of the dataset so removing them was justified.

#### 1. Set up appropriate hypothesis test
* **H0:** ρ=0 **H1:** ρ≠0
* **Null hypothesis:** There is no correlation between the *Number of Discharges* and the *Excess Readmission Ratio* (correlation coefficient is 0, ρ=0)
* **Alternate hypothesis:** There is correlation between the *Number of Discharges* and the *Excess Readmission Ratio* (correlation coefficient is not 0, ρ≠0)

To test this hypothesis we will simulate the collection of sample data by bootstrapping (resampling the original sample with replacement to get a new sample of the same size) and calculate the correlation coefficient for each bootstrap (calculate the bootstrap replicate). From now on I will refer to the correlation coefficient by ρ. Once we have all ρ, we will calculate their mean and standard deviation. By the law of large numbers, this mean will be a very good estimator of the actual population ρ, however since we are performing a test under the null hypothesis, we need to use it's prescribed ρ value of 0 for the Z value calculation. The standard deviation (given the large sample size) will act as an unbiased estimate of the population standard deviation of ρ, we will also be able to use this as our ρ distribution standard deviation (after diving by √n, n being sample size). The last important point to make is that since we are testing correlation, we need to perform a pairs bootstrap, this means that we maintain (*Number of Discharges, Excess Readmission Ratio*) pairs.

#### 2. Compute and report the observed significance value (or p-value).

```python
#Under the null hypothesis the proposed correlation coeffiecient is 0.
corr_coef_proposed=0
#Here we calculate the observed correlation coefficient from our original data (clean_hospital_read_df dataframe).
observed_corr_coef, p_value_precalculated=stats.pearsonr(clean_hospital_read_df["Number of Discharges"], 
                                         clean_hospital_read_df["Excess Readmission Ratio"])

number_of_observations=len(clean_hospital_read_df["Number of Discharges"]) #Number of rows in cleaned up dataset.

corr_coef_values=np.empty(10000) #We are simulating 10,000 samples, so we need an array to hold 10,000 ρ values.

for i in range(10000):
   #From an array of dataframe indeces, choose the required amount for a bootstrap sample 
   simulated_sample_indeces=np.random.choice(range(number_of_observations), number_of_observations)
   #Select the appropriate pairs of values from the same observation (pairs bootstrap) and calculate correlation coefficient.
   simulated_corr_coef, junk_p_value=stats.pearsonr(clean_hospital_read_df.loc[simulated_sample_indeces,"Number of Discharges"], 
                                         clean_hospital_read_df.loc[simulated_sample_indeces, "Excess Readmission Ratio"])
   corr_coef_values[i]=simulated_corr_coef
```

Here I completed the pairs bootstrap and computed 10,000 ρ values and stored them in the *corr_coef_values* array. The following output is the distribution of those ρ values and includes the observed ρ value as well as the mean ρ value of the ρ sampling distribution. As mentioned before, by the law of large numbers, for a large number of sample statistics the mean of the sample means is very close to the true population ρ.

```python
#Calculating the mean of generated ρ values.
corr_coef_mean=np.mean(corr_coef_values)

plt.hist(corr_coef_values, bins=100, normed=True, color="black", label="Simulated ρ values")
plt.xlabel("ρ value") 
plt.ylabel("Probability")
plt.axvline(x=observed_corr_coef, color="red", label="Observed ρ value") 
plt.axvline(x=corr_coef_mean, color="blue", label="Mean ρ value") 
plt.legend(loc="upper left")
plt.show()
print("Observed corr coef: %f, Mean corr coef: %f" % (observed_corr_coef, corr_coef_mean))
```

As hypothesized, our *observed ρ value* is very closely approximated by the *mean ρ value* leading to the conclusion that our observed correlation is very close to the true population correlation coefficient. We will now back this up mathematically.

```python
corr_coef_std=np.std(corr_coef_values, ddof=1)/np.sqrt(len(corr_coef_values))
z_value=(observed_corr_coef-corr_coef_proposed)/corr_coef_std
print("Z value:", z_value)
```

#### 3. Report statistical significance for  α = .01

After consulting a Z table we find -2.325 to be the critical Z value for an α=.01 significance level. Any Z value lower than that has an associated p value lower than 0.01 and is statistically significant. Given that the associated Z value (~-841) of our observed correlation coefficient under the null hypothesis is much lower than the critical value of -2.325, **we reject the null hypothesis in favor of the alternate hypothesis and conclude that the correlation coefficient ρ is not 0**. Furthermore, having computed a large number of sample statistics from bootstrapping (10,000 of them) by the law or large numbers we can be fairly confident that the true population correlation coefficient is very very close to -0.09.

#### 4.  Discuss statistical significance and practical significance. Do they differ here? How does this change your recommendation to the client?
The result of the hypothesis test in terms of statistical significance is that the *excess readmissions ratio* correlates negatively with the *number of discharges*, however the strength of that negative correlation is only ~-0.09. This means that, in terms of practical significance, this finding does not provide a useful recommendation in itself. It's also important to note that even if the strength of the correlation would have been strong, correlation does not imply causation, thus taking action to increase or decrease the *number of discharges* would not necessarily decrease the number of hospitals with an *excess readmissions ratio*. So given that the data is statistically significant, but not practically significant, we conclude that *'they do differ here'*. So the recommendation would be to gather more parameters which might correlate with *excess readmission ratio*. Another recommendation for reducing the *excess readmission ratio* (or in other words, reducing the number of readmissions*) would be to introduce stricter requirements for a person to be discharged. This would naturally reduce the number of readmissions.
